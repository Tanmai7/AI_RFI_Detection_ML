```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/features.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
label_map = {
    0: "Clean",
    1: "Narrowband",
    2: "Broadband",
    3: "Impulsive"
}

df["Class"] = df["Label"].map(label_map)
df
```

```python
df["Class"].value_counts().plot(kind="bar")
plt.title("Class Distribution of RF Signals")
plt.xlabel("Signal Type")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x="Class", y="Variance", data=df)
plt.title("Variance Distribution Across Signal Types")
plt.show()
```

```python
sns.barplot(x="Class", y="Peak_Frequency", data=df)
plt.title("Peak Frequency Comparison")
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=["number"])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")
plt.show()
```
